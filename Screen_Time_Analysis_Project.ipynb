### Screen Time Analysis is the task of analyzing and creating a report on which applications and websites are used by the user for how much time.

### Apple devices have one of the best ways of creating a screen time report.Screenshot 2023-07-16 154558.png



For the task of screen time analysis, I will be using the dataset that contains data about the following:

Date
Usage of Applications
Number of Notifications from Applications
Number of times apps opened

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
data = pd.read_csv("/content/Screentime-App-Details.csv")
data.head()
```

```python
#Checking if the data set contains any null values or not

data.isnull().sum()
```

Above result shows that the data set does not contain any null values.

```python
#Looking at the descriptive statistics of the data

data.describe()
```

To analyse the screen time, checking the amount of usage of app

```python
figure = px.bar(data_frame = data,
                x = "Date",
                y = "Usage",
                color = "App",
                title = "App Usage")

figure.show()
```

Now analysing the number of notifications from the apps

```python
figure = px.bar(data_frame = data,
                x = "Date",
                y = "Notifications",
                color = "App",
                title = "Notifications of app")

figure.show()
```

Analysing the number of times the app has been opened

```python
figure = px.bar(data_frame = data,
                x = "Date",
                y = "Times opened",
                color = "App",
                title = "Times an application was opened")

figure.show()
```

Checking the realtionship between number of notifications and the amount of usage, since we generally open the app when we recieve notifications from the app.

```python
figure = px.scatter(data_frame=data,
                    x = "Notifications",
                    y = "Usage",
                    size = "Notifications",
                    trendline = "ols",
                    title = "Relationship between Number of notifications and usage")
figure.show()
```

### Since, there is a linear relationship between the number of notifications and the amount of usage.

### It means that more notifications result in more use of smartphones.
